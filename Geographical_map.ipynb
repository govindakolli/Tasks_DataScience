**Generating Geographical Heatmap**

Create a geographical heatmap using Matplotlib and geographical data (e.g., basemap).

```python
#impoting libraries
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

#Creating some synthetic data (e.g., temperature)
#Define the grid of longitudes and latitudes
lon = np.linspace(-180, 180, 360)
lat = np.linspace(-90, 90, 180)
lon_grid, lat_grid = np.meshgrid(lon, lat)

# Create synthetic data (e.g., temperature values)
temperature_data = np.sin(np.radians(lat_grid)) * np.cos(np.radians(lon_grid)) * 30 + 30

# Create a figure with a geographical projection
fig, ax = plt.subplots(figsize=(12, 6), subplot_kw={'projection': ccrs.PlateCarree()})

# Add coastlines and gridlines
ax.coastlines()
ax.gridlines(draw_labels=True)

# Define the colormap
cmap = plt.get_cmap('coolwarm')

# Plot the heatmap
heatmap = ax.pcolormesh(lon_grid, lat_grid, temperature_data, cmap=cmap, shading='auto')

# Add a color bar
cbar = plt.colorbar(heatmap, ax=ax, orientation='vertical', pad=0.02)
cbar.set_label('Temperature (Â°C)')

# Add title
ax.set_title('Geographical Heatmap of Temperature')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

#grid of longitudes and latitudes for India
lon = np.linspace(68, 97, 100)
lat = np.linspace(8, 37, 100)
lon_grid, lat_grid = np.meshgrid(lon, lat)

#Create synthetic data (e.g., temperature values)
temperature_data = 30 + 10 * np.sin((lat_grid - 20) / 10) * np.cos((lon_grid - 80) / 10)

# Create a figure with a geographical projection
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw={'projection': ccrs.PlateCarree()})

# Add features and coastlines
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.STATES)
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.LAKES, facecolor='blue', alpha=0.5)
ax.add_feature(cfeature.RIVERS)

# Plot the heatmap
heatmap = ax.pcolormesh(lon_grid, lat_grid, temperature_data, cmap='coolwarm', shading='auto')

# Add a color bar
cbar = plt.colorbar(heatmap, ax=ax, orientation='vertical', pad=0.02)
cbar.set_label('Temperature (Â°C)')

# Set the title
ax.set_title('Geographical Heatmap of India')

# Set the extent to focus on India
ax.set_extent([68, 97, 8, 37], crs=ccrs.PlateCarree())

# Show the plot
plt.show()
```
