**Multiple Categorical Variables using Crosstab in Pandas**

*Pandas Contingency Table*

A contingency table, also known as a cross-tabulation, is a table that summarizes the relationship between two or more categorical variables in a dataset. In Pandas, you can create a contingency table using the crosstab() function.

Syntax


*pd.crosstab(index, columns, values=None, rownames=None, colnames=None, aggfunc=None, margins=False, margins_name='All', dropna=True, normalize=False)*

```python
import pandas as pd
```

```python
# data
data = {
    "Names" : ["Ajay", "Sruthi", "Venu", "Anjali", "Govind", "Sowmya",
               "Ajay", "Sruthi", "Venu", "Anjali", "Govind", "Sowmya" ],
    "Gender" : ["Male", "Female", "Male", "Female", "Male", "Female",
                "Male", "Female", "Male", "Female", "Male", "Female"],
    "Exam": ["Semester 1","Semester 1","Semester 1","Semester 1","Semester 1","Semester 1",
            'Semester 2','Semester 2','Semester 2','Semester 2','Semester 2','Semester 2'],
    'Result':['Pass','Pass','Fail','Pass','Fail','Pass','Pass','Fail','Fail','Pass','Pass','Fail']

}
```

```python
df = pd.DataFrame(data)
df
```

```python
#2 way cross rable
my_crosstab = pd.crosstab(df.Exam, df.Result , margins = True)
my_crosstab
```

```python
#Rename the index and columns

my_crosstab.columns = ["Fail" , "Pass" , "rowtotal"]
my_crosstab.index= ["Semester 1","Semester 2","coltotal"]
my_crosstab
```

```python
#3 way cross table
pd.crosstab([df.Exam, df.Gender], df.Result , margins = True)
```
